```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("ret_data_capm.csv")
```

```python
df.head()
```

```python
# Trái phiếu chính phủ theo tháng
df["TB1"] = (1 + df["TB3MS"]) ** (1/3) - 1

# Return của tesla trừ trái phiếu
df["R_TSLA"] = df["TSLA"] - df["TB1"]

# Return của walmart trừ trái phiếu
df["R_WMT"] = df["WMT"] - df["TB1"]

# Return của thị trường trừ trái phiếu
df["R_M"] = df["SP500"] - df["TB1"]
df.head()
```

Cau 1.b

```python
for x in ["TSLA", "WMT"]:
    plt.figure(figsize = (10,5))
    sns.regplot(data = df, x = "SP500", y = x)
    plt.title(f"Scatter plot of SP500 and {x}")
    plt.show()
```

Chắc là nó tuyến tính.

Cau 1.c

Mô hình tesla

```python
from statsmodels.formula.api import ols

# Biến phụ thuộc & độc lập
y = df["R_TSLA"]
x = df["R_M"]

# Chạy hồi quy
formula = "y ~ x"
model = ols(formula = formula, data = df).fit()

model.summary()
```

p value < 0.05 chắc ổn rồi.

Market model Tesla: R_it - RF_t = 1.8097*(RM_t - RF_t) + 0.0333

Mô hình walmart 

```python
# Biến phụ thuộc & độc lập
y = df["R_WMT"]
x = df["R_M"]

# Chạy hồi quy
formula = "y ~ x"
model = ols(formula = formula, data = df).fit()

model.summary()
```

Market model Walmart: R_it - RF_t = 0.504*(RM_t - RF_t) + 0.0047
